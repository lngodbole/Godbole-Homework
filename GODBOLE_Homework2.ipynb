We are going to study the survival rate of passengers on titanic and what variables affected survival.

***1. Load the dataset in `titanic.xls`. It contains data on all the passengers that travelled on the Titanic. Show the head of 10 rows (10 pts)***

```python
# libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# settings
pd.set_option('display.max_rows', None) 
pd.set_option('display.max_columns', 15)

df = pd.read_excel('titanic.xls')

print(df.shape)
```

```python
df.head(10)
```

```python
df.tail(10)
```

*** 2. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (20 pts)*** 

```python
femaleTotal = df['sex'].value_counts()['female'] # counting total amount of females in df
maleTotal = df['sex'].value_counts()['male'] # counting total amount of males in df

femaleTotal+maleTotal # ensuring that the total matches the shape of the entire dataframe


df2 = df.groupby('sex').sum() # grouping by sex and summing up values to get total survived

df2['total'] = [femaleTotal, maleTotal]

     
df2['SurvivalRate'] = df2['survived'] / df2['total']

toDrop = ['sibsp', 'parch', 'fare', 'body', 'age']

df2.drop(toDrop, inplace = True, axis = 1)

print(df2.to_markdown())

```

*** 3. Calculate the same proportion, but by pclass and sex. (30 pts)*** 

```python
# Calculations for pclass and sex

df['count'] = 1 # adding a count variable to the original dataframe with 1 in every row.  
                #That way when I do my groupbys, I will get the total individuals in each group
    
df3 = df.groupby(['pclass', 'sex']).sum().reset_index() # grouping by sex and summing up values to get total survived

    
df3['SurvivalRate'] = df3['survived'] / df3['count']


df3.drop(toDrop, inplace = True, axis = 1)
df3 = df3.sort_values(by = 'SurvivalRate', ascending=False)


print(df3.to_markdown())

```

*** 4. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, pclass and sex. (30 pts)***

```python
# Creating the age tiers

tiers = [
    (df['age'] < 14),
    (df['age'] >= 14) & (df['age']<=20),
    (df['age']  >= 21) & (df['age']<=64),
    (df['age'] >= 65)
]
     
names = ['children (under 14)', 'adolescents (14-20)', 'adult (21-64)', 'senior (65+)']


df['Age Category'] = np.select(tiers, names)

df.head(50)
```

```python
# Calculations for pclass, age category, and sex

df['count'] = 1 # adding a count variable to the original dataframe with 1 in every row.  
                #That way when I do my groupbys, I will get the total individuals in each group
    
df4 = df.groupby(['pclass', 'sex', 'Age Category']).sum().reset_index() # grouping by sex and summing up values to get total survived

    
df4['SurvivalRate'] = df4['survived'] / df4['count']
df4['Age Category'].replace(to_replace='0', value="unknown", inplace = True)

df4.drop(toDrop, inplace = True, axis = 1)

#print(df4.to_markdown(tablefmt="grid"))

df4 = df4.sort_values(by = 'SurvivalRate', ascending=False)

df4['pclass'] = df4['pclass'].astype(str)


#df4['Labels'] = 'Class: ' + df4.pclass + ' , Sex: ' + df.sex + ' , Age Category: ' + df['Age Category']
df4.sort_values("SurvivalRate")

#df4.head(100)
print(df4.to_markdown())
```
